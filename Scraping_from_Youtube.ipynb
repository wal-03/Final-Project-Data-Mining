## Code for Youtube API Comments

```python
import googleapiclient.discovery

api_key = "your_api_key"


def search_and_get_comments(search_query, max_comments_per_video=1000):
    """
    Searches for videos based on a query and retrieves comments from the top results.

    Args:
        search_query: The search query to use for finding YouTube videos.
        max_comments_per_video: The maximum number of comments to retrieve from each video (default 10).

    Returns:
        A tuple containing:
            - A dictionary where keys are video titles and values are lists of comments from that video.
            - The total number of comments retrieved across all videos.
            - The number of videos retrieved from the search.
    """

    # Create a YouTube service object
    youtube = googleapiclient.discovery.build("youtube", "v3", developerKey=api_key)

    # Search for videos
    search_response = youtube.search().list(
        part="snippet",
        q=search_query,
        maxResults=1000  # Adjust as needed
    ).execute()

    # Extract video IDs and titles from search results
    video_titles_and_ids = {}
    for item in search_response['items']:
        video_id = item['id']['videoId']
        video_title = item['snippet']['title']
        video_titles_and_ids[video_id] = video_title

    # Create an empty dictionary to store comments
    video_comments = {}
    total_comments = 0  # Initialize a variable to store total comments
    num_videos_fetched = len(video_titles_and_ids)  # Initialize variable for fetched videos

    # Loop through the retrieved video IDs
    for video_id, video_title in video_titles_and_ids.items():
        try:
            # Get comments for the current video
            request = youtube.commentThreads().list(
                part="snippet",
                videoId=video_id,
                maxResults=max_comments_per_video
            )
            response = request.execute()

            # Extract comments from the response
            video_comments[video_title] = []  # Create an empty list for this video's comments
            for item in response['items']:
                try:
                    comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
                    video_comments[video_title].append(comment)
                    total_comments += 1  # Increment total comments for each comment retrieved
                except KeyError:  # Handle cases where topLevelComment might be missing
                    pass

        except Exception as e:  # Handle potential errors during video retrieval
            print(f"Error retrieving comments for video ID {video_id}: {e}")

    return video_comments, total_comments, num_videos_fetched


# Example usage
search_query = "kenaikan ump 6,5%"

all_comments, total_comments, num_videos_fetched = search_and_get_comments(search_query)

print(f"\nTotal comments retrieved: {total_comments}")
print(f"\nNumber of videos fetched: {num_videos_fetched}")
print(f"\nComments per video:\n")
for comments in all_comments.values():
    for comment in comments:
        print(comment)
```
